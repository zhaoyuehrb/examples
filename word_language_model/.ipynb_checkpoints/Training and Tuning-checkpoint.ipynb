## Imports, Loading and Pre-processing taken from example code

```python
import argparse
import time
import math
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.onnx
import torch.optim as optim

import data
import model
```

```python
## args ##
dat_path = './data/wikitext-2'
batch_size = 4000
seed = 1111
bptt = 35
window = 5
lr = 0.001
torch.manual_seed(seed)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
corpus = data.Corpus(dat_path)
```

```python
def batchify(data, bsz):
    # Work out how cleanly we can divide the dataset into bsz parts.
    nbatch = data.size(0) // bsz
    # Trim off any extra elements that wouldn't cleanly fit (remainders).
    data = data.narrow(0, 0, nbatch * bsz)
    # Evenly divide the data across the bsz batches.
    data = data.view(bsz, -1).t().contiguous()
    return data.to(device)

eval_batch_size = batch_size
train_data = batchify(corpus.train, batch_size)
val_data = batchify(corpus.valid, eval_batch_size)
test_data = batchify(corpus.test, eval_batch_size)
```

```python
ntokens = len(corpus.dictionary)
```

```python
print(train_data.size())
print(val_data.size())
print(test_data.size())
print(ntokens)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
def repackage_hidden(h):
    """Wraps hidden states in new Tensors, to detach them from their history."""

    if isinstance(h, torch.Tensor):
        return h.detach()
    else:
        return tuple(repackage_hidden(v) for v in h)
```

```python
def get_batch(source, i):
    seq_len = min(bptt, len(source) - 1 - i)
    data = source[i:i+seq_len]
    target = source[i+1:i+1+seq_len].view(-1)
    return data, target


def get_batch_window(source, i, window): # only 1 after last is target
    seq_len = window
    data = source[i:i+seq_len]
    target = source[i+1+seq_len].view(-1)
    return data.t(), target
```

## Feedforward Model defination

### When enabling input-output embedding sharing, hidden_size is forced to be the same as embed_dim

```python
## Model

class FNNModel(nn.Module):
    def __init__(self,vocab_size,window = 5,embed_dim = 300, share = False, hidden_size = 300):
        super(FNNModel,self).__init__()
        self.window = window
        self.embed_dim = embed_dim
        self.vocab_size = vocab_size
        self.hidden_size = hidden_size
        self.share = share
        if share:
            self.hidden_size = embed_dim
        else:
            self.linear2 = nn.Linear(hidden_size, vocab_size, bias = False)
        
        self.input_embed = nn.Embedding(vocab_size, embed_dim)

        self.linear1 = nn.Linear(window * embed_dim, hidden_size)
        
    
    def forward(self, inputs):
        lookup = self.input_embed(inputs)
#        print(lookup.shape)
        concat = lookup.view(-1, self.window * self.embed_dim)
#        print(concat.shape)
        mid = torch.tanh(self.linear1(concat))
#        print(mid.shape)
        if self.share:
            logits = torch.matmul(mid, self.input_embed.weight.t())
        else:
            logits = self.linear2(mid)
#        print(logits.shape)
        return logits

```

```python
model = FNNModel(ntokens,share=True)
model.to(device)
```

## Training and Evaluation adapted from example codes.

For evaluation, only the denominator is changed, because the number of words evaluated is different from RNN model.

For training, loss and perplexity are printed every 200 iteration, and the last iteration.

```python
def evaluate(data_source):
    # Turn on evaluation mode which disables dropout.
    model.eval()
    total_loss = 0.
    ntokens = len(corpus.dictionary)
    with torch.no_grad():
        for i in range(0, data_source.size(0) - 1 - window):
            data, targets = get_batch_window(data_source, i,window)
            output = model(data)
            total_loss += criterion(output, targets).item()
    avg_loss = total_loss / (len(data_source) - 1 - window)
    return avg_loss, math.exp(avg_loss)
```

```python

epoch = 0
def train():
    # Turn on training mode which enables dropout.
    model.train()
    count = 0
    total_loss = 0.
    start_time = time.time()
    ntokens = len(corpus.dictionary)
    for batch, i in enumerate(range(0, train_data.size(0) - window-1)):
        count += 1
        data, targets = get_batch_window(train_data, i,window)
        # Starting each batch, we detach the hidden state from how it was previously produced.
        # If we didn't, the model would try backpropagating all the way to start of the dataset.
        model.zero_grad()
        output = model(data)
#        print(output.shape)
#        print(targets.shape)
#        print(output,targets)
        loss = criterion(output, targets)
        loss.backward()

        optimizer.step()

        total_loss += loss.item()

        if batch % 200 == 0 and batch > 0:
            cur_loss = total_loss / 200
            elapsed = time.time() - start_time
            print('| epoch {:3d} | {:5d}/{:5d} batches | lr {:03.4f} | ms/batch {:5.2f} | '
                    'loss {:5.2f} | ppl {:8.2f}'.format(
                epoch, batch, len(train_data), lr,
                elapsed * 1000 / 200, cur_loss, math.exp(cur_loss)))
            total_loss = 0
            start_time = time.time()
            count = 0
#         if dry_run:
#             break

    cur_loss = total_loss / count
    elapsed = time.time() - start_time
    print('| epoch {:3d} | {:2d}/{:2d} batches | lr {:03.4f} | ms/batch {:5.2f} | '
            'loss {:5.2f} | ppl {:8.2f}'.format(
        epoch, batch, len(train_data), lr,
        elapsed * 1000 / 200, cur_loss, math.exp(cur_loss)))
    return cur_loss,math.exp(cur_loss)
```

## Hyperparameter tuning and recording

Hyperparameters tuned includes 'whether to share input-output embedding', 'window_size', 'learning rate'. Here embedding size is forced to be the same as hidden layer size due to resource constraint. This is not necessary when not sharing input-output. The whole tuning takes around 24 hours on a RTX2070S.

Training/evaluation/test loss and perplexity are recorded at the end of each epoch. Only evaluation loss is used to select hyperparameters. Test loss and perplexity is only used for final reporting.

```python
import pandas as pd
import numpy as np
```

```python
train_log = pd.DataFrame(columns = ['embed_dim','window','lr','share','epoch','loss','ppl','type'])
epochs = 40
for share in [False,True]:
    for window in range(2,5):
        for embed_dim in range(100,501,200):
            for lr in [0.001,0.0001]:
                del model
                model = FNNModel(ntokens,window=window,share=share,embed_dim=embed_dim,hidden_size=embed_dim)
                model.to(device)
                optimizer = optim.Adam(model.parameters(), lr = lr)
                for epoch in range(epochs):
                    el,lp = train()
                    train_log = train_log.append({'embed_dim': embed_dim,
                                      'window': window,
                                     'lr': lr,
                                     'share': share,
                                     'epoch': epoch,
                                     'loss': el,
                                     'ppl': lp,
                                     'type': 'train'},ignore_index=True)                    
                    el,lp = evaluate(val_data)
                    print('eval_loss: {:5.2f} ; eval_ppl : {:5.2f}'.format(el,lp))
                    train_log = train_log.append({'embed_dim': embed_dim,
                                      'window': window,
                                     'lr': lr,
                                     'share': share,
                                     'epoch': epoch,
                                     'loss': el,
                                     'ppl': lp,
                                     'type': 'eval'},ignore_index=True)
                    el,lp = evaluate(test_data)
                    print('test_loss: {:5.2f} ; test_ppl : {:5.2f}'.format(el,lp))
                    train_log = train_log.append({'embed_dim': embed_dim,
                                      'window': window,
                                     'lr': lr,
                                     'share': share,
                                     'epoch': epoch,
                                     'loss': el,
                                     'ppl': lp,
                                     'type': 'test'},ignore_index=True)
                train_log.to_csv('train_log.csv',index=False)
```
